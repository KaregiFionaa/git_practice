# Data Preparation

Here we ensure ensure that the dataset is clean, structured, and ready for exploration and modeling. 
The key step is to load the dataset into a pandas DataFrame for analysis.

```python
#importation of libraries
import pandas as pd
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_style("whitegrid")

# Load the CSV file with specified encoding and data type
df = pd.read_csv('AviationData.csv', encoding='ISO-8859-1', dtype='unicode')

# Display the first few rows
df.head()
```

Columns: Identify the columns in the dataset and their data types.

Summary Statistics: Generate summary statistics to understand the distribution of numerical variables.

```python
# Display column names and data types
df.info()

# Generate summary statistics
df.describe(include='all')
```

Data Volume
Size: Determine the number of rows and columns in the dataset.

```python
# Number of rows and columns

rows, columns = df.shape
print(f"The dataset contains {rows} rows and {columns} columns.")
```

```python
print("Column Names:", df.columns.tolist())
```

### Filter by Specific Criteria

To make our analysis more manageable, we will filter the dataset to focus on the following relevant columns:
- **Event.Id**: Unique identifier for each event.
- **Number.of.Engines**: The number of engines on the aircraft.
- **Injury.Severity**: Severity of injuries reported.
- **Total.Fatal.Injuries**: Total number of fatal injuries.
- **Total.Serious.Injuries**: Total number of serious injuries.
- **Total.Uninjured**: Number of individuals not injured.
- **Aircraft.Category**: Category of the aircraft.
- **Purpose.of.flight**: The purpose for which the aircraft was used.
- **Weather.Condition**: Weather conditions during the event.
- **Make**: Manufacturer of the aircraft.
- **Model**: Model of the aircraft.

We will filter the dataset to keep only these columns.

```python
# We will select only relevant columns
relevant_columns = [
    'Event.Id', 
    'Number.of.Engines', 
    'Injury.Severity', 
    'Total.Fatal.Injuries', 
    'Total.Serious.Injuries',
    'Total.Uninjured', 
    'Aircraft.Category',
    'Purpose.of.flight', 
    'Weather.Condition', 
    'Make', 
    'Model'
]

# Filter the DataFrame
df_filtered = df[relevant_columns]

# Display the filtered DataFrame
df_filtered.head()
```

### Checking for duplicates

Next, we will check for any duplicate rows in the filtered dataset.

```python
# Check for duplicate rows
duplicates = df_filtered.duplicated().sum()
print("Number of duplicate rows:", duplicates)

# Remove duplicates if any
if duplicates > 0:
    df_filtered = df_filtered.drop_duplicates()
    print("Duplicates removed. New shape of the DataFrame:", df_filtered.shape)
```

### Check Data Types

We will now check the data types of each column in our filtered dataset to ensure they are appropriate for our analysis.

```python
# Display data types of each column
df_filtered.dtypes
```

```python
#Check for any outliers in the numerical columns

df_filtered.describe()
```

### Overview of Filtered Data

The filtered DataFrame now contains only the relevant columns, which will allow us to focus our analysis on key factors related to aircraft safety and performance.

Next, we will clean the filtered dataset by handling missing values and ensuring that the data types are appropriate for our analysis.

```python
#we'll check for missing values in the filtered Df
missing_values = df_filtered.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Check for non-numeric values
df_filtered['Total.Fatal.Injuries'].unique()
```

```python
df_filtered['Total.Fatal.Injuries'] = pd.to_numeric(df_filtered['Total.Fatal.Injuries'], errors='coerce')
df_filtered.dropna(subset=['Total.Fatal.Injuries', 'Total.Serious.Injuries'], inplace=True)
df_filtered['Total.Fatal.Injuries'] = df_filtered['Total.Fatal.Injuries'].astype(int)
```

```python
df_filtered['Number.of.Engines'] = pd.to_numeric(df_filtered['Number.of.Engines'], errors='coerce')
df_filtered['Total.Uninjured'] = pd.to_numeric(df_filtered['Total.Uninjured'], errors='coerce')
```

```python
df_filtered['Total.Serious.Injuries'] = pd.to_numeric(df_filtered['Total.Serious.Injuries'], errors='coerce')
df_filtered['Total.Serious.Injuries'].isnull().sum() # Count NaNs

# Fill NaNs with 0 or another value as needed
df_filtered['Total.Serious.Injuries'].fillna(0, inplace=True)
```

```python
df_filtered.dtypes
```

```python
# Getting the values present in the Aircraft.Category
df_filtered["Aircraft.Category"].value_counts(dropna= False)
```

```python
#company is interested in only purchasing airplanes so we'll drop all other values in the Aircraft.Category

df_studied =df_filtered[df_filtered['Aircraft.Category'] == 'Airplane'].reset_index(drop=True)
df_studied.info()
```

```python
#checking for the first few rows

df_studied.head()
```

```python
#To drop the Aircraft.Category since they're all Airplanes
df_studied= df_studied.copy()
df_studied.drop(["Aircraft.Category"], axis=1,inplace = True)
df_studied.head(10)
```

##  Analyze Total Fatal and Serious Injuries by Aircraft Make and Model

In this step, we will group the data by `Make` and `Model` to analyze the total number of fatal and serious injuries. This will help us identify which aircraft have the least number of fatal incidents.
We will then group the DataFrame by `Make` and `Model`, and aggregate the total fatal injuries, serious injuries, and uninjured counts.

```python
# Group by Make and Model into a new column "Aircraft_Model"
df_studied['Aircraft_Model'] = df_studied['Make'] + ' ' + df_studied['Model']

# Drop the original "Make" and "Model" columns
df_studied = df_studied.drop(['Make', 'Model'], axis=1)

# Display the structure of the updated DataFrame
df_studied.info()
```

```python
df_studied['Aircraft_Model'] = df_studied['Aircraft_Model'].str.upper()
df_studied.info()
```

```python
unique_models = df_studied['Aircraft_Model'].unique()
unique_models
```

```python
# Group by Aircraft_Model and aggregate total fatal injuries
model_injury_summary = df_studied.groupby('Aircraft_Model')['Total.Fatal.Injuries'].sum().reset_index()
model_injury_summary
```

```python
accident_counts = df_studied['Aircraft_Model'].value_counts().reset_index()
accident_counts.columns = ['Aircraft_Model', 'Accident_Count']
accident_counts
```

```python
engine_comparison = df_studied.groupby(['Aircraft_Model', 'Number.of.Engines']).size().unstack()
engine_comparison
```

```python
injury_severity_comparison = df_studied.groupby('Aircraft_Model')['Injury.Severity'].value_counts().unstack()
injury_severity_comparison
```

```python
weather_comparison = df_studied.groupby(['Aircraft_Model', 'Weather.Condition']).size().unstack()
weather_comparison
```

```python
purpose_comparison = df_studied.groupby(['Aircraft_Model', 'Purpose.of.flight']).size().reset_index(name='Accident_Count')
purpose_comparison
```

```python
uninjured_summary = df_studied.groupby('Aircraft_Model')['Total.Uninjured'].sum().reset_index()
uninjured_summary
```

## Visualization

In this part of our notebook, we'll create visualizations to compare the safety records of different aircraft models. This will lay the groundwork for covering further analysis , allowing us to identify the airplane model with the lowest risk.

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df_studied, x='Aircraft_Model', y='Total.Fatal.Injuries', inner='quartile', color='blue')
plt.title('Distribution of Total Fatal Injuries by Aircraft Model', fontsize=20)
plt.xlabel('Aircraft Model', fontsize=14)
plt.ylabel('Total Fatal Injuries', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

The accidents were non fatal.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_filtered, 
              x='Purpose.of.flight', 
              order=df_filtered['Purpose.of.flight'].value_counts().index, 
              color="blue")

plt.title('Total Serious Injuries by Purpose of Flight', fontsize=20)
plt.xlabel('Purpose of Flight', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.xticks(rotation=90)  
plt.show()
```

Personal Aircrafts had the most amount of accidents.

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_studied,
             x='Number.of.Engines',  
             y='Total.Fatal.Injuries',  
             estimator=sum,  
             color="blue")

plt.title('Total Fatal Injuries by Number of Engines', fontsize=20)
plt.xlabel('Number of Engines', fontsize=14)
plt.ylabel('Total Fatal Injuries', fontsize=14)
plt.xticks(rotation=45) 
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Total Fatal Injuries were on the rise when the Airplane was using 1 engine.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_studied, 
              x='Weather.Condition', 
              order=df_studied['Weather.Condition'].value_counts().index, 
              color="blue")

plt.title('Count of Aviation Incidents by Weather Condition', fontsize=20)
plt.xlabel('Weather Condition', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

It's durng VMC weather that majority of the accidents occured.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_studied, 
              x='Aircraft_Model', 
              hue='Injury.Severity', 
              order=df_studied['Aircraft_Model'].value_counts().index, 
              color="blue")

plt.title('Count of Aviation Accidents by Aircraft Model and Injury Severity', fontsize=20)
plt.xlabel('Aircraft Model', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)  
plt.legend(title='Injury Severity', loc='upper right')
plt.show()
```

Most accidents happened when it was 1 engine.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_studied, x='Number.of.Engines', y='Total.Serious.Injuries')
plt.title('Total Serious Injuries by Number of Engines', fontsize=20)
plt.xlabel('Number of Engines', fontsize=14)
plt.ylabel('Total Serious Injuries', fontsize=14)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Serious injuries occured when there were 2 engines.

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_studied, 
                 x='Weather.Condition', 
                 y='Total.Serious.Injuries', 
                 hue='Weather.Condition', 
                 palette='Set2', 
                 alpha=0.7)

plt.title('Total Serious Injuries by Weather Condition', fontsize=20)
plt.xlabel('Weather Condition', fontsize=14)
plt.ylabel('Total Serious Injuries', fontsize=14)
plt.xticks(rotation=45)  # Rotate x labels for better visibility
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Serious injuries were recorded during the VMC weather which refers to weather conditions that allow a pilot to fly visually, meaning they can rely on their eyesight to maintain safe separation from terrain and other aircraft.

## Summary of Findings

The analysis of aviation incidents provided valuable insights into the relationships between various factors, including aircraft model, weather conditions, engine types, number of engines, and purpose of flight.

a.) Weather Conditions: The analysis revealed that VMC weather conditions significantly contribute to the frequency and severity of injuries. Incidents occurring in poor weather conditions were associated with a notable increase in both serious and fatal injuries.

b.) Engine Types and Number of Engines: The examination of engine types and the number of engines revealed distinct patterns in injury rates. Aircraft with 1 or 2 engine configurations may have different performance characteristics that affect safety. Additionally, the number of engines appeared to play a role in the severity of incidents.

3.)Purpose of Flight: The personal flight also influenced injury patterns. private flights displayed varying levels of risk.

4.)Injury Patterns: The distributions of serious and fatal injuries varied across the factors analyzed, indicating the importance of contextual elements in understanding aviation safety.

5.)Aircraft Model Impact: Certain airplane models exhibited higher average fatal injuries, indicating a correlation between specific designs and safety outcomes.
